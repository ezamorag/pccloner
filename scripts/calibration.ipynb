```python
import pynput 
from keymapping_win11 import mapping

def on_press(key):
    keyname = str(key)
    equik = mapping.get(keyname, key)
    print(key, ' => ', equik)

def on_release(key):
    print('released', key)

listener = pynput.keyboard.Listener(on_press=on_press) #, on_release=on_release)
listener.start()
```

```python
listener.stop()
```

```python
from pynput.keyboard import Key, Controller
import time 

kController = Controller()

for k in Key.__dict__['_member_map_'].keys(): 
    print('here', k)
    kController.press(Key.__dict__[k])
    kController.release(Key.__dict__[k])
    time.sleep(0.5)
```
