```python
import os, re
import subprocess

def run_command(command):
    """Runs a shell command and prints the output."""
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    output, error = process.communicate()
    if output:
        print(output.decode())
    if error:
        print(error.decode())

def getversion():
    with open("setup.py", "r") as f:
        content = f.read()
    match = re.search(r"version='([^']+)'", content)
    if match:
        version = match.group(1)
    else:
        print("Version not found.")
    return version 

def uninstallingpccloner():
    # Run the command, sending 'y\n' (yes + newline) as input
    process = subprocess.Popen("pip uninstall pccloner", shell=True, stdin=subprocess.PIPE, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True)
    output, error = process.communicate(input='y\n') 
    # Print the output and any errors
    print(output)
    if error:
        print(error)
        
# Make sure that yuo have the environment below: 
# conda create --name x python==3.10
# conda activate x
# pip install pyinstaller
# pip install --upgrade build
```

### Routine for producing the pip package and the ubuntu app

```python
base_dir = '/home/ezamorag/GitHub/pccloner'
os.chdir(base_dir)  # Replace with your actual path
print('Building ...')
run_command('python -m build')
print('Downloading PIP package ...')
run_command('twine upload --repository pypi dist/*')

print(f'Installing the latest pccloner version {getversion()} ...')
uninstallingpccloner()
run_command('pip install pccloner==' + getversion())
print('Generating executable for ubuntu ...')
run_command('pyinstaller -F scripts/testing.py -n pccloner_applinux_' + getversion())
run_command('pyinstaller -F scripts/pccollector.py -n pccollector_applinux_' + getversion())
```
