```python
from pynput import keyboard
from pynput.keyboard import Key
from pynput import mouse

blockedknames = ['ctrl', 'ctrl_l', 'ctrl_r', 
                 'alt', 'alt_l', 'alt_r', 'alt_gr',
                 'shift', 'shift_r', 'shift_l']
blockedkeys =[Key.__dict__[kname] for kname in blockedknames]
previouskey = 0

def on_press(key):
    global previouskey

    if not ((key in blockedkeys) and (key == previouskey)):
        keystring = str(key).strip("'")
        px, py = mouse.Controller().position
        print(px, py, f'pressed {keystring}', 'previous', previouskey)
    previouskey = key


def on_release(key):
    global previouskey
    if key == previouskey:
        previouskey = 0

listener = keyboard.Listener(on_press=on_press, on_release=on_release)
listener.start()
```

```python
listener.stop()
```
