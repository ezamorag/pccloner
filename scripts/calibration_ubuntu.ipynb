```python
import pynput 

mapping = {
'<269025048>': 'f4', 
'<269025062>': 'f6',  
'<65027>': 'alt_gr', 
'<65056>': 'tab',   # when shift+tap
"['´']": None, # "['´']": '´',  # tilde   
"['¨']": None, # "['¨']": '¨',  # tilde con shift_r
'"\'"': "'",   
}

def on_press(key):
    key = mapping.get(str(key), key)
    keystring = str(key).strip("'") if str(key) != "'" else str(key)
    print('pressed', keystring)

def on_release(key):
    print('released', key)

listener = pynput.keyboard.Listener(on_press=on_press, on_release=on_release)
listener.start()
```

```python
listener.stop()
```
